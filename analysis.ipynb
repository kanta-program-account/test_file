```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'C:\\Users\kyabe\OneDrive\デスクトップ\Python\Works\\gi_train_mm10.csv'
df = pd.read_csv(file_path)
df['event_day'] = pd.to_datetime(df['event_day'])
df['event_day'] = df['event_day'].dt.strftime('%d')

df.info()
df.describe()
```

```python
#product_idと商品対応表及び売り上げ数
df1 = df.dropna(subset = ['product_id', 'product_name', 'buy_flag'])
df1.info()

purchase_by_product = df1.groupby(['product_id', 'product_name'], as_index = False)['buy_flag'].sum()
print(purchase_by_product)

#商品ごとの売り上げ数
print('商品ごとの売り上げ:')
plt.bar(purchase_by_product['product_id'], purchase_by_product['buy_flag'])
plt.title('purchase_by_product')
plt.xlabel('product_id')
plt.ylabel('purchase')
plt.xticks(rotation = 90)
plt.show()

#product_id 0~60の拡大版棒グラフ
print('product_idが0-60の商品における売り上げ:')
plt.bar(purchase_by_product['product_id'], purchase_by_product['buy_flag'])
plt.title('purchase_by_product')
plt.xlabel('product_id')
plt.ylabel('purchase')
plt.xticks(rotation = 90)
plt.xlim(0, 59.5)
plt.show()

#売り上げtop5
top5 = purchase_by_product[purchase_by_product['buy_flag'] > 120]
top5 = top5.sort_values('buy_flag', ascending = False)
top5['rank'] = range(1, 6)
print('全体の売り上げtop5:')
print(top5)
```

```python
#性別における商品ごとの売り上げ
purchase_by_gender = df.groupby('gender', as_index = False)['buy_flag'].sum()
print('男女別の総売り上げ')
print(purchase_by_gender)

#男性の商品別売り上げ
purchase_by_man = df[df['gender'] == 'man']
purchase_by_man = purchase_by_man.groupby(['product_id', 'product_name'], as_index = False)['buy_flag'].sum()
print('男性の商品別売り上げ:')
plt.bar(purchase_by_man['product_id'], purchase_by_man['buy_flag'])
plt.title("man's_purchase")
plt.xlabel('product_id')
plt.ylabel('purchase')
plt.xticks(rotation = 90)
plt.show()

top5_man = purchase_by_man[purchase_by_man['buy_flag'] > 50]
top5_man = top5_man.sort_values('buy_flag', ascending = False)
top5_man['rank'] = range(1, 6)
print('男性の売り上げtop5:')
print(top5_man)

#女性の商品別売り上げ
purchase_by_woman = df[df['gender'] == 'woman']
purchase_by_woman = purchase_by_woman.groupby(['product_id', 'product_name'], as_index = False)['buy_flag'].sum()
print('女性の商品別売り上げ:')
plt.bar(purchase_by_woman['product_id'], purchase_by_woman['buy_flag'])
plt.title("woman's_purchase")
plt.xlabel('product_id')
plt.ylabel('purchase')
plt.xticks(rotation = 90)
plt.show()

top5_woman = purchase_by_woman[purchase_by_woman['buy_flag'] > 70]
top5_woman = top5_woman.sort_values('buy_flag', ascending = False)
top5_woman['rank'] = range(1, 6)
print('女性の売り上げtop5:')
print(top5_woman)
```

```python
#イベント時間経過と売り上げの関係
df['in_time'] = pd.to_datetime(df['in_time'])
df['start_hour'] = df['in_time'].dt.hour
df['out_time'] = pd.to_datetime(df['out_time'])
df['finish_time'] = df['out_time'].dt.hour

df2 = df.dropna(subset = ['in_time', 'time_duration', 'product_id', 'product_name', 'buy_flag'])
df2.info()

purchase_by_hour_all = df2.groupby('start_hour', as_index = False)['buy_flag'].sum()

purchase_by_hour = df2.groupby(['start_hour', 'gender'], as_index = False)['buy_flag'].sum()
purchase_by_hour_man = purchase_by_hour[purchase_by_hour['gender'] == 'man']
purchase_by_hour_woman = purchase_by_hour[purchase_by_hour['gender'] == 'woman']

plt.plot(purchase_by_hour_all['start_hour'], purchase_by_hour_all['buy_flag'], marker='o', label='all')
plt.plot(purchase_by_hour_man['start_hour'], purchase_by_hour_man['buy_flag'], marker='s', label='man')
plt.plot(purchase_by_hour_woman['start_hour'], purchase_by_hour_woman['buy_flag'], marker='v', label='woman')

plt.title('purchase_by_hour')
plt.xlabel('hour')
plt.ylabel('purchase')
plt.legend()
plt.show()
```

```python
#イベント開催時間の長さと購入数の関係
#イベントの開催時間の表記と単位が不明の為、解析不可。
bins = [i for i in range(0, 380, 10)]
df3 = df2.dropna(subset = ['time_duration', 'buy_flag'])

df3['time_duration_category'] = pd.cut(df3['time_duration'], bins = bins, right = False)


purchase_by_TimeDuration = df3.groupby('time_duration_category', as_index = False)['buy_flag'].sum()

plt.hist(purchase_by_TimeDuration['buy_flag'], bins = bins)
plt.title('Purchase by Time Duration')
plt.xlabel('Time Duration Category')
plt.ylabel('Purchase')
plt.xticks(rotation=90)
plt.show()
```

```python
#年齢と購入数の相関　scatter
purchase_by_age = df.groupby('age', as_index = False)['buy_flag'].sum()
plt.scatter(purchase_by_age['age'], purchase_by_age['buy_flag'])
plt.title('purchase_by_age')
plt.xlabel('age')
plt.ylabel('purchase')
plt.show()

purchase_by_age['age'] = purchase_by_age['age'].str.replace("Over", '')
purchase_by_age['age'] = purchase_by_age['age'].str.replace("'s", '')
purchase_by_age['age'] = purchase_by_age['age'].astype(float)

corr = purchase_by_age.corr()
print(f'相関行列:\n{corr}')
```
